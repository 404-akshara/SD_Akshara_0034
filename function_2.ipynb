```python
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a + b

```

```python
for i in fibonacci(5):
  print(i)
```

```python
fib=fibonacci(10)
```

```python
print(fib)
```

```python
def count_test(n):
  count = 1
  while count<=n:
    yield count
    count=count+1
```

```python
for i in count_test(3):
  print(i)
```
